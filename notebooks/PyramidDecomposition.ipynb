```python
%pylab inline
from flowpm import linear_field, lpt_init, nbody, cic_paint

import tensorflow as tf
gpu = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpu[0], True)

from scipy.interpolate import InterpolatedUnivariateSpline as iuspline

klin = np.loadtxt('../flowpm/data/Planck15_a1p00.txt').T[0]
plin = np.loadtxt('../flowpm/data/Planck15_a1p00.txt').T[1]

ipklin = iuspline(klin, plin)
```

```python
import flowpm

stages = np.linspace(0.1, 1.0, 10, endpoint=True)

initial_conditions = flowpm.linear_field(64,         # size of the cube
                                         50,         # Physical size of the cube
                                         ipklin,      # Initial powerspectrum
                                         batch_size=1)

# Sample particles
state = flowpm.lpt_init(initial_conditions, a0=0.1)   

# Evolve particles down to z=0
final_state = flowpm.nbody(state, stages, 64)         

# Retrieve final density field
final_field = flowpm.cic_paint(tf.zeros_like(initial_conditions), final_state[0])
```

```python
def get_bspline_kernel(num_channels, dtype=tf.float32, order=4):
  """Creates a b-spline kernel of requested order.
  Args:
    num_channels: The number of channels of the image to filter.
    dtype: The type of an element in the kernel.
  Returns:
    A tensor of shape `[x, x, x, num_channels, num_channels]`.
  """
  if order == 8:
    kernel = np.array(( 1., 8., 28., 56., 70., 56., 28., 8., 1.), dtype=dtype.as_numpy_dtype())
  elif order == 6:
    kernel = np.array(( 1., 6., 15., 20., 15., 6., 1.), dtype=dtype.as_numpy_dtype())
  elif order==2:
    kernel = np.array(( 1., 2., 1.), dtype=dtype.as_numpy_dtype())
  else:
    kernel = np.array(( 1., 4., 6., 4., 1.), dtype=dtype.as_numpy_dtype())
  kernel = np.einsum('ij,k->ijk', np.outer(kernel, kernel), kernel)
  kernel /= np.sum(kernel)
  kernel = kernel[:, :, :, np.newaxis, np.newaxis]
  return tf.constant(kernel, dtype=dtype) * tf.eye(num_channels, dtype=dtype)

def _downsample(cube, kernel, antialias=True):
  """Downsamples the image using a convolution with stride 2.
  """
  cube = tf.nn.conv3d(input=cube, filters=kernel, strides=[1, 2, 2, 2, 1], padding="SAME")
  if antialias:
    k = get_bspline_kernel(kernel.shape[-1], order=2)
    cube = tf.nn.conv3d(input=cube, filters=k, strides=[1, 1, 1, 1, 1], padding="SAME")
  return cube
  
def _upsample(cube, kernel, output_shape=None):
  """Upsamples the image using a transposed convolution with stride 2.
  """
  if output_shape is None:
    output_shape = tf.shape(input=cube)
    output_shape = (output_shape[0], output_shape[1] * 2, output_shape[2] * 2,
                    output_shape[3] * 2, output_shape[4])
  return tf.nn.conv3d_transpose(
      cube,
      kernel * 2.0**3,
      output_shape=output_shape,
      strides=[1, 2, 2, 2, 1],
      padding="SAME")

def _build_pyramid(cube, sampler, num_levels):
  """Creates the different levels of the pyramid.
  """
  kernel = get_bspline_kernel(1, dtype=cube.dtype)
  levels = [cube]
  for _ in range(num_levels):
    cube = sampler(cube, kernel)
    levels.append(cube)
  return levels

def _split(cube, kernel):
  """Splits the image into high and low frequencies.
  This is achieved by smoothing the input image and substracting the smoothed
  version from the input.
  """
  low = _downsample(cube, kernel)
  high = cube - _upsample(low, kernel, tf.shape(input=cube))
  return high, low

def downsample(cube, num_levels, name=None):
  """Generates the different levels of the pyramid (downsampling).
  """
  with tf.name_scope("downsample"):
    cube = tf.convert_to_tensor(value=cube)
    return _build_pyramid(cube, _downsample, num_levels)

def merge(levels, name="merge"):
  """Merges the different levels of the pyramid back to an image.
  """
  with tf.name_scope(name):
    levels = [tf.convert_to_tensor(value=level) for level in levels]
    cube = levels[-1]
    kernel = get_bspline_kernel(tf.shape(input=cube)[-1], dtype=cube.dtype)
    for level in reversed(levels[:-1]):
      cube = _upsample(cube, kernel, tf.shape(input=level)) + level
    return cube

def split(cube, num_levels, name="split"):
  """Generates the different levels of the pyramid.
  """
  with tf.name_scope(name):
    cube = tf.convert_to_tensor(value=cube)

    kernel = get_bspline_kernel(tf.shape(input=cube)[-1], dtype=cube.dtype)
    low = cube
    levels = []
    for _ in range(num_levels):
      high, low = _split(low, kernel)
      levels.append(high)
    levels.append(low)
    return levels

def upsample(cube, num_levels, name="upsample"):
  """Generates the different levels of the pyramid (upsampling).
  """
  with tf.name_scope(name):
    cube = tf.convert_to_tensor(value=cube)
    return _build_pyramid(cube, _upsample, num_levels)
```

## Computing Forces on Multi-Desolution Decomposition

```python
# Compute forces on both fields
def force(field):    
    shape = field.get_shape()
    batch_size, nc = shape[1], int(shape[2])
    kfield = flowpm.utils.r2c3d(field)
    kvec = flowpm.kernels.fftk((nc, nc, nc), symmetric=False)
    
    
    lap = tf.cast(flowpm.kernels.laplace_kernel(kvec), tf.complex64)
    fknlrange = flowpm.kernels.longrange_kernel(kvec, 0)
    kweight = lap * fknlrange
    pot_k = tf.multiply(kfield, kweight)
    
    f = []
    for d in range(3):
      force_dc = tf.multiply(pot_k, flowpm.kernels.gradient_kernel(kvec, d))
      forced = flowpm.utils.c2r3d(force_dc)
      f.append(forced)
    
    return tf.stack(f, axis=-1)
```

```python
# Split field into short range and large scale components
levels = split(field, 1)
```

```python
force_levels = [force(levels[0][...,0]), force(levels[1][...,0])*2]
```

```python
rec = merge(force_levels)
```

```python
# Direct force computation on input field
dforce = force(field[...,0])
```

```python
figure(figsize=(15,5))
subplot(131)
imshow(final_field[0].numpy().sum(axis=1))
title('Input')
subplot(132)
imshow(force_levels[0].numpy()[0,...,0].sum(axis=1))
title('short range forces')
subplot(133)
imshow(force_levels[1].numpy()[0,...,0].sum(axis=1));
title('l2')
title('long range forces')
```

```python
figure(figsize=(17,5))
subplot(131)
imshow(rec[0].numpy().sum(axis=1)[...,0]);
title('Multi-Grid Force Computation')
colorbar();
subplot(132)
imshow(dforce[0].numpy().sum(axis=1)[...,0]);
title('Direct Force Computation')
colorbar();
subplot(133)
imshow((rec - dforce).numpy()[0,8:-8,8:-8,8:-8].sum(axis=1)[...,0]); 
title('Residuals'); 
colorbar();
```

```python
def force_mr(l0, l1):    
    ker = get_bspline_kernel(1, order=4)
    
    kl0 = flowpm.utils.r2c3d(l0)
    kl1 = flowpm.utils.r2c3d(l1)
    
    shape = l0.get_shape()
    batch_size, nc = shape[1], int(shape[2])
    
    kvec0 = flowpm.kernels.fftk((nc, nc, nc), symmetric=False)
    kvec1 = [k/2 for k in  flowpm.kernels.fftk((nc//2, nc//2, nc//2), symmetric=False)]
    
    
    lap0 = tf.cast(flowpm.kernels.laplace_kernel(kvec0), tf.complex64)
    fknlrange0 = flowpm.kernels.longrange_kernel(kvec0, 0)
    kweight0 = lap0 * fknlrange0
    
    lap1 = tf.cast(flowpm.kernels.laplace_kernel(kvec1), tf.complex64)
    fknlrange1 = flowpm.kernels.longrange_kernel(kvec1, 0)
    kweight1 = lap1 * fknlrange1
        
    pot_k0 = tf.multiply(kl0, kweight0)
    pot_k1 = tf.multiply(kl1, kweight1)
    
    f0 = []
    f1 = []
    for d in range(3):
        f0.append(flowpm.utils.c2r3d(tf.multiply(pot_k0, flowpm.kernels.gradient_kernel(kvec0, d))))
        f1.append(flowpm.utils.c2r3d(tf.multiply(pot_k1, flowpm.kernels.gradient_kernel(kvec1, d))))
    
    df = []
    for g,f in zip(f0,f1):
        # this is the downsampled low resolution field
        f = tf.expand_dims(f, -1)
        f = _upsample(f, ker)
        
        # This is the high resolution details component
        g = tf.expand_dims(g, -1)
        _low = _downsample(g, ker)
        g = g - _upsample(_low, ker)
        
        df.append(f[...,0]+g[...,0])
    
    return (tf.stack(f0, axis=-1), tf.stack(f1, axis=-1)), tf.stack(df, axis=-1)
```

```python
flevels, frec = force_mr(l0[...,0],l1[...,0])
```
